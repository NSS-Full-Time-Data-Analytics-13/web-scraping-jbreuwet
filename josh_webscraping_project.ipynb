## #3 Extracting Job Descriptions

```python
desc_url = 'https://realpython.github.io/fake-jobs/jobs/senior-python-developer-0.html'
```

```python
response2 = requests.get(desc_url)
```

```python
soup2 = BS(response2.text)
```

```python
soup2
```

```python
soup2.find('div', {'class':'content'}).text
```

```python
def scrape(x):
    scrape_response = requests.get(str(x))
    scrape_soup = BS(scrape_response.text)
    return(scrape_soup.find('div', {'class':'content'}).text)
```

```python
job_descriptions = jobs['URL'].apply(scrape)
```

```python
job_descriptions = job_descriptions.to_frame()
```

```python
job_descriptions.head()
```

```python
job_descriptions['URL'] = job_descriptions['URL'].str.replace('\n','')
```

```python
job_descriptions.head()
```

```python
job_descriptions = job_descriptions.rename(columns = {'URL':'Job Description'})
```

```python
job_descriptions.head()
```

```python
jobs = pd.concat([jobs, job_descriptions], axis=1)
```
